## Maple Leaves Ltd. 
#### is a start-up company which makes herbs from different types of plants and its leaves. Currently the system they use to classify the trees which they import in a batch is quite manual. A labourer from his experience decides the leaf type and subtype of plant family. They have asked us to automate this process and remove any manual intervention from this process.Objective: To classify the plant leaves by various classifiers from different metrics of the leaves and to choose the best classifier for future reference.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plants = pd.read_csv('/Users/umangpandya/Desktop/Edureka Project/574_cert_proj_dataset_v3.0/train.csv')
```

```python
plants.shape
```

```python
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
    
```

```python
plants
```

```python
plantnames = plants.iloc[:,1]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
plantnames = le.fit_transform(plantnames)
```

```python
plantnames = pd.DataFrame(plantnames)
```

```python
species = pd.DataFrame(plants.species)
```

```python
ids = pd.DataFrame(plants.id)
```

```python
plants
```

```python
plants.drop(columns = ['id','species'],axis = 1,inplace = True)
```

```python
plants = pd.concat([plants,plantnames],axis = 1)
```

```python
plants.isna().sum()
```

```python
plants = pd.concat([plants, species],axis = 1)
```

```python
plants.rename(columns ={0:'Label_Codes','species' : 'Species'}, inplace = True )
```

```python
plants = pd.concat([plants,ids],axis = 1)
```

```python
plants.head()
```

```python
y = plants.iloc[:,-3].values
x = plants.iloc[:,:192].values
```

```python
from sklearn.preprocessing import StandardScaler
scx = StandardScaler()
x = scx.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.2)
```

I am not going to perform Logistic Regression since looking at the categories its vast and cannot be solved through Logistic regression

# KNearestNeighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
knn.fit(x_train,y_train)
#Test Data
y_pred = knn.predict(x_test)
y_pred



```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy')
dt.fit(x_train, y_train)
#Test Data
y_pred = dt.predict(x_test)
y_pred  
```

```python
accuracy_score(y_test, y_pred)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion = 'entropy')
rf.fit(x_train, y_train)
#Test Data
y_pred = rf.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
y_pred = nb.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

# Support Vector Machine

```python
from sklearn.svm import SVC
sv = SVC (kernel = 'linear')
sv.fit(x_train, y_train)
```

```python
y_pred = sv.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
sv = SVC (kernel = 'sigmoid')
sv.fit(x_train, y_train)
```

```python
y_pred = sv.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
sv = SVC (kernel = 'rbf')
sv.fit(x_train, y_train)
```

```python
y_pred = sv.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
